Get all color codes available for matplotlib for visualization

```python
def plot_all_cmaps():
    N_ROWS, N_COLS = 8, 7 # 13, 13 <-- for all in one figure 
    HEIGHT, WIDTH = 7, 14

    cmap_ids = plt.colormaps()
    n_cmaps = len(cmap_ids)
    
    print(f'mpl version: {mpl.__version__},\nnumber of cmaps: {n_cmaps}')
    
    index = 0
    while index < n_cmaps:
        fig, axes = plt.subplots(N_ROWS, N_COLS, figsize=(WIDTH, HEIGHT))
        for row in range(N_ROWS):
            for col in range(N_COLS):
                ax = axes[row, col]
                cmap_id = cmap_ids[index]
                cmap = plt.get_cmap(cmap_id)
                mpl.colorbar.ColorbarBase(ax, cmap=cmap,
                                          orientation='horizontal')
                ax.set_title(f"'{cmap_id}', {index}", fontsize=8)
                ax.tick_params(left=False, right=False, labelleft=False,
                               labelbottom=False, bottom=False)
                
                last_iteration = index == n_cmaps-1
                if (row==N_ROWS-1 and col==N_COLS-1) or last_iteration:
                    plt.tight_layout()
                    #plt.savefig('colormaps'+str(index)+'.png')
                    plt.show()
                    if last_iteration: return
                index += 1

plot_all_cmaps()
```

Lets calculate the highest increase in total population within a ten(10) year period 

```python
# get the dataset with column :total census
ten_year_max_increase = df_sum_total

# get the needed column 
max_10 = ten_year_max_increase[['Total Census']] 
```

```python
max_10
```

```python
#since we need the highest increase in total population within a 
#ten year period,lets drop rows 2015 and 2022

max_10.drop(['2015','2022'],inplace = True)
max_10
```

```python
#now after dropping rows 2015 and 2022,lets calculate the difference

diff = max_10.diff()
diff
```

```python
# after getting the difference, lets sort in descending order
#to get the final results

diff.sort_values(['Total Census'],ascending  = False, inplace = True)
```

```python
diff
```

```python
# lets plot the data to get a clear picture of the results
diff.plot(kind = 'bar',y = 'Total Census',figsize= (12,7),color = colors_list)
plt.title('Highest increase in Total population for 10 years period')
plt.ylabel('Diff.Total Population(m)')
plt.xlabel('Ten year interval')

x = ['1970','1980','1990','2000','2010','2020',]
y = diff['Total Census']
for i in range(len(x)):
    plt.text(i,y[i],y[i],ha="center",va=("bottom"))
    
#plt.savefig('Highest in total population for ten year interval.jpg', bbox_inches='tight')
    
```
